```python
import tensorflow as tf

x_train = [1,2,3]
y_train = [1,2,3]

W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

hypothesis = x_train * W + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - y_train))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2000):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```

```python
X = tf.placeholder(tf.float32, shape=[None])
Y = tf.placeholder(tf.float32, shape=[None])

WW = tf.Variable(tf.random_normal([1]), name='weight')
bb = tf.Variable(tf.random_normal([1]), name='bias')

hHypothesis = WW*X + bb

Ccost = tf.reduce_mean(tf.square(hHypothesis - Y))
```

```python
Ooptimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
Ttrain = Ooptimizer.minimize(Ccost)
```

```python
session = tf.Session()
session.run(tf.global_variables_initializer())

for i in range(2001):
    cost_val, W_val, b_val, _ = session.run([Ccost, WW, bb, Ttrain], feed_dict={X:[1,2,3], Y:[1,2,3]})
    if i % 20 == 0:
        print (i, cost_val, W_val, b_val)
```
